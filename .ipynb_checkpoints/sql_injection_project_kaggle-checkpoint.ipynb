## 1. Load SQL Injection Datasets

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 500)
```

```python
data_1 = pd.read_csv("sqli.csv", encoding="UTF-16 LE")
data_2 = pd.read_csv("sqliv2.csv", encoding="UTF-16 LE")
data_3 = pd.read_csv("SQLiV3.csv", encoding="UTF-8")
```

## 2. Merge DataSets

```python
# Merge all dataframes
data = pd.concat([data_1, data_2, data_3], axis=0, ignore_index=True)
```

## 3. Data Exploration

```python
# Show a random sample of 5 rows from the data
data.sample(5)
```

```python
data.info()
```

## 4. Data Wrangling

### 4.1. Clean Erroneous Label

```python
data.Label.unique()[:10]
```

We need to clean some data points which contain labels which are neither 0 nor 1

```python
correct_labels = [1, 0, '1', '0']
```

```python
data = data[data['Label'].isin(correct_labels)]
```

```python
# Check if every incorrect labels have been dropped
data.Label.unique()
```

```python
# map label for consistency
data['Label'] = data['Label'].replace({'1': 1, '0': 0})
```

### 4.2. Drop Unnamed Columns

```python
# Drop Unnamed columns
data = data.drop(columns=['Unnamed: 2', 'Unnamed: 3'], axis=1)
```

```python
# Confirm Unnamed columns have been dropped
data.info()
```

### 4.3. Drop NaNs

```python
data = data.dropna()
```

```python
# Confirm all NaNs have been drop
data.info()
```

## 5. Data Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(data['Sentence'])
y = data['Label']
```

## 6. Split Data into Training and Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## 7. Define Evaluation Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
```

```python
def plot_confusion_matrix(conf_matrix):
    plt.figure()
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.show()
```

```python
def plot_roc_curve(classifier, X_test, y_test):
    y_prob = classifier.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-AUC Curve')
    plt.legend(loc='lower right')
    plt.show()
```

## 8. Model Building

```python
def train_and_evaluate_classifier(classifier, X_train, X_test, y_train, y_test):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)

    return accuracy, precision, recall, f1, conf_matrix
```

```python
def evaluate_classifier(classifier, X_train, X_test, y_train, y_test):
    accuracy, precision, recall, f1, conf_matrix = train_and_evaluate_classifier(classifier, X_train, X_test, y_train, y_test)

    print(f'Accuracy: {accuracy:.2f}')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1-Score: {f1:.2f}')

    plot_confusion_matrix(conf_matrix)
    plot_roc_curve(classifier, X_test, y_test)
```

### 8.1. Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
# Naive Bayes Classifier
nb_classifier = MultinomialNB()
evaluate_classifier(nb_classifier, X_train, X_test, y_train, y_test)
```

### 8.2. Ensemble Model

```python
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
# Define base classifier
nb_classifier_base = MultinomialNB()

# Define final-classifier
svm_classifier = SVC(probability=True, random_state=42)
```

```python
# Build Ensemble Model
ensemble_classifier = StackingClassifier(
    estimators=[('nb', nb_classifier_base)],
    final_estimator=svm_classifier,
    stack_method='auto',
)

# Evaluate Stacking Ensemble
evaluate_classifier(ensemble_classifier, X_train, X_test, y_train, y_test)
```

## 9. Save Model

```python
import joblib
```

```python
model = {
    'vectorizer': vectorizer,
    'nb_classifier': nb_classifier,
    'ensemble_classifier': ensemble_classifier
}
```
